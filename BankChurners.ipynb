```python
df['Attrition_Flag'] = df.Attrition_Flag.apply(lambda x : 1 if x == 'Existing Customer' else 0)
df.Attrition_Flag
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Gender'] = le.fit_transform(df.Gender)
df['Education_Level'] = le.fit_transform(df.Education_Level)
df['Marital_Status'] = le.fit_transform(df.Marital_Status)
df['Income_Category'] = le.fit_transform(df.Income_Category)
df['Card_Category'] = le.fit_transform(df.Card_Category)
df.head()

#I think it is better to use label encoding to avoid high cardinality
```

```python
labels = 'Male', 'Female'
sizes = [df.Gender[df['Gender']==1].count(), df.Gender[df['Gender']==0].count()]
explode = (0, 0.1)
fig1, ax1 = plt.subplots(figsize=(10, 8))
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')
plt.title("Gender Percentage", size = 20)
plt.show()
```

```python
labels = 'Churned', 'Retained'
sizes = [df.Attrition_Flag[df['Attrition_Flag']==0].count(), df.Attrition_Flag[df['Attrition_Flag']==1].count()]
explode = (0, 0.1)
fig1, ax1 = plt.subplots(figsize=(10, 8))
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')
plt.title("Proportion of customer churned and retained", size = 20)
plt.show()
```

```python
labels = 'Married', 'Unmarried'
sizes = [df.Marital_Status[df['Marital_Status']==1].count(), df.Marital_Status[df['Marital_Status']==0].count()]
explode = (0, 0.1)
fig1, ax1 = plt.subplots(figsize=(10, 8))
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')
plt.title("Proportion of Married and Unmarried customers", size = 20)
plt.show()
```

```python
num = df_num.corrwith(df['Attrition_Flag']).sort_values(ascending=False)
num
```

```python
def plot_corr(scores):
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title("Correlation Scores")


plt.figure(dpi=100, figsize=(8, 5))
plot_corr(num)
```

```python
X = df.iloc[:,1:]
y = df.iloc[:,:1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.20,random_state=0) 
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train,y_train)
```

```python
svc.score(X_test,y_test)
```

```python
svc.score(X_train,y_train)
```

```python
from sklearn import metrics
pred = svc.predict(X_test)
confusion_matrix = metrics.confusion_matrix(y_test, pred)
confusion_matrix
```

```python
from sklearn.metrics import classification_report,precision_score,recall_score
print(classification_report(y_test,pred))
```

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()

scaledX = scale.fit_transform(X)

print(scaledX)
```

```python
X_train_S, X_test_S, y_train, y_test = train_test_split(scaledX,y,test_size=.20,random_state=0)
```

```python
svc = SVC()
svc.fit(X_train_S,y_train)
```

```python
svc.score(X_test_S,y_test)
```

```python
svc.score(X_train_S,y_train)
```

```python
from sklearn import metrics
pred = svc.predict(X_test_S)
confusion_matrix = metrics.confusion_matrix(y_test, pred)
confusion_matrix
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train_S, y_train)
```

```python
rf.score(X_test_S, y_test)
```

```python
rf_predict = rf.predict(X_test_S)
print(classification_report(y_test,rf_predict))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
y_prob = rf.predict_proba(X_test_S)[:, 1]

# Calculate the false positive rate, true positive rate, and thresholds for the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

# Calculate the AUC score
auc_score = roc_auc_score(y_test, y_prob)

# Plot the ROC curve
plt.plot(fpr, tpr, label=f'AUC={auc_score:.3f}')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random Guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive')
```
